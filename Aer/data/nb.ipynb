```python
# Visualisation des données : nombre de zones d'échantillonages et nombre d'échantillons par zone

import pandas as pd

# === 1. Charger le fichier CSV ===
# Remplace 'donnees.csv' par le chemin de ton fichier
df = pd.read_csv("indice_pollen_bretagne_epci.csv")

# === 2. Nombre total de zones différentes ===
nb_zones = df['lib_zone'].nunique()
print(f"Nombre de zones différentes : {nb_zones}")

# === 3. Nombre de mesures par zone ===
mesures_par_zone = df.groupby('lib_zone')['date_ech'].count().reset_index()
mesures_par_zone.rename(columns={'date_ech': 'nombre_mesures'}, inplace=True)

print("\nNombre de mesures par zone :")
print(mesures_par_zone)
```
